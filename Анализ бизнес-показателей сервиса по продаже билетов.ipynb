# Маркетинговое исследование для сервиса по продаже билетов

Наш заказчик - сервис "Яндекс.Афиша". На основе логов сервера с данными о посещениях сайта, статистике всех заказов за этот период и статистики рекламных расходов мы должны помочь маркетологам снизить расходы: отказаться от невыгодных источников трафика и перераспределить бюджет.

В нашем распоряжении три датасета.

Файл **visits_log.csv** хранит лог сервера с информацией о посещениях сайта, **orders_log.csv** — информацию о заказах, а **costs.csv** — информацию о расходах на маркетинг.

**Структура visits_log.csv**

- Uid — уникальный идентификатор пользователя,
- Device — категория устройства пользователя,
- Start Ts — дата и время начала сессии,
- End Ts — дата и время окончания сессии,
- Source Id — идентификатор источника перехода на сайт.

**Структура orders_log.csv**

- Uid — уникальный идентификатор пользователя,
- Buy Ts — дата и время заказа,
- Revenue — сумма заказа.

**Структура costs.csv**

- source_id — идентификатор рекламного источника,
- dt — дата проведения рекламной кампании,
- costs — расходы на эту кампанию.

 4. Сделать выводы по всем метрикам и выдать рекомендации по применению наиболее эффективных рекламных источников

```python
# Берем из старого проекта функцию для оценки количества пропущенных значений во всем датафрейме или в отдельном столбце

def check_na(df, column_to_check = None):    
    if column_to_check is None: check_list = df.columns
    else: 
        check_list = [column_to_check]        
    for column in check_list:        
        null_values = df[column].isna().sum()
        overall_values = df[column].count() + null_values
        percentage = round(null_values / overall_values * 100,2)
        print(f'Пропущенных значений {column}: {null_values} из {overall_values} ({percentage}%)')

df_names = ['visits', 'orders', 'costs']

for df_name in df_names:
    print('Анализ пропущенных значений в датафрейме ', df_name)
    check_na(globals()[df_name])
    print()
```

## Предообработка данных

```python
# переименовываем столбцы

for df_name in df_names:
    globals()[df_name].columns = globals()[df_name].columns.str.lower()
visits = visits.rename(columns = {'end ts': 'end_ts', 'source id': 'source_id', 'start ts': 'start_ts'})
orders = orders.rename(columns = {'buy ts': 'buy_ts'})
```

```python
# меняем формат данных в столбцах с датами и типами источников

visits['end_ts'] = pd.to_datetime(visits['end_ts'])
visits['start_ts'] = pd.to_datetime(visits['start_ts'])
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])
costs['dt'] = pd.to_datetime(costs['dt'])

# проверим, не "прицепились" ли к данным заказы или визиты за следующие периоды
print(visits['start_ts'].sort_values(ascending=False).head(10))
print(orders['buy_ts'].sort_values(ascending=False).head(10))
```

```python
# создаем столбцы с датами и месяцам
visits['month']  = visits['start_ts'].astype('datetime64[M]')
visits['week']  = visits['start_ts'].astype('datetime64[W]')
visits['date']  = visits['start_ts'].astype('datetime64[D]')

orders['month'] = orders['buy_ts'].astype('datetime64[M]')
orders['date'] = orders['buy_ts'].astype('datetime64[D]')

costs['month'] = costs['dt'].astype('datetime64[M]')
```

```python
# создаем и применяем функцию, определяющую когорты пользователей

def define_cohorts(df, time_input):
    cohort_month = df.groupby('uid')[time_input].min().astype('datetime64[M]')
    cohort_month.name = 'cohort_month'
    df = df.merge(cohort_month, on='uid', how='left')
    df['cohort_lifetime'] = ((df[time_input].astype('datetime64[M]') - df['cohort_month']) / np.timedelta64(1, 'M')).apply(round)
    cohort_size = df.groupby('cohort_month')['uid'].nunique().reset_index()
    cohort_size = cohort_size.rename(columns=({'uid': 'cohort_size'}))
    df = df.merge(cohort_size, on='cohort_month', how='left')
    return df

visits = define_cohorts(visits, 'start_ts')
orders = define_cohorts(orders, 'buy_ts')
```

### Частота сессий

Определим, сколько раз в день среднестатистический пользователь заходит на сайт и изучим изменение метрики во времени.

```python
sessions_per_user = visits.groupby(['date']).agg({'uid': ['count','nunique']})
sessions_per_user.columns = ['n_sessions', 'n_users']
sessions_per_user['n_sessions'] = sessions_per_user['n_sessions'] / sessions_per_user['n_users']
display(sessions_per_user['n_sessions'].describe())

plt.figure(figsize=(12, 6))
plt.title('Среднее количество сессий в день')
plt.grid(True)
plt.plot(sessions_per_user['n_sessions'], label='Количество сессий')
plt.margins(x=0)
plt.legend()
plt.show()
```

Как мы видим на графике, кроме единичных случаев, частота посещаемости сайта пользователями находится в диапазоне **1.05-1.1 раз в день**. Устраним выбросы, ограничив данные стандартным диапазоном в 1.5 межквартильных интервала.

```python
# функция для устранения выбросов, пригодится в дальнейшем

def remove_outliers(df, column, k):
    q25, q75 = np.percentile(df[column], 25), np.percentile(df[column], 75) 
    iqr = q75 - q25
    lower, upper = q25 - iqr*1.5, q75 + iqr*1.5
    return(df[(df[column] >= lower) & (df[column] <= upper)])

sessions_per_user = remove_outliers(sessions_per_user, 'n_sessions', 1.5)
display(sessions_per_user['n_sessions'].describe())

plt.figure(figsize=(12, 6))
plt.title('Среднее количество сессий в день')
plt.grid(True)
plt.plot(sessions_per_user['n_sessions'], label='Количество сессий')
plt.margins(x=0)
plt.legend()
plt.show()
```

Согласно очищенным от выбросов данным, пользователи посещают сайт **от 1.03 до 1.13 раз в сутки**. Среднее значение и медиана практически равны и составляют **1.08 раз в сутки**. Как мы видим, отклонение минимально, ярко выраженного сезонного фактора в этом показателе - не обнаружено. Иными словами, частота ежедневных посещений сайта, в отличие от количества пользователей, не зависит от времени года.

### Продолжительность сессий

Выявим продолжительность типичной пользовательской сессии.

```python
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
print(visits['session_duration_sec'].describe())

plt.figure(figsize=(12, 6))
plt.grid()
plt.hist(visits['session_duration_sec'], bins=200)
plt.show()
```

Как мы видим, 4-й квартиль находится в диапазоне от 14 минут до практически 24 часов. При очевидной невозможности суточной продолжительности сессий трактуем верхнюю часть диапазона как мусорные значения. Возможно, это просто незакрытые неактивные сессии, либо ошибка в логах. Избавимся от выбросов, ограничив исследуемые данные 95-м перцентилем.

```python
session_durations = visits.query('session_duration_sec <= 2280')['session_duration_sec']
print(session_durations.describe())
```

Продолжительность средней сессии в очищенных от выбросов данных - **от 0 (т.е. до 60 секунд) до 38 минут**. Для оценки данной метрики разделим сессии на категории и оценим их долю:
 - Очень короткие сессии: **меньше 1 минуты**
 - Короткие сесии: **от 1 до 5 минут**
 - Сессии средней продолжительности: **от 50 до 20 минут**
 - Длительные сессии: **более 20 минут**

```python
duration_values = (visits['session_duration_sec'].value_counts().reset_index()
    .rename(columns = {'index':'value', 'session_duration_sec': 'n_values'}))
durations_df = pd.DataFrame(columns = ['type', 'n_values'])
duration_types = {
        (0,59): 'Менее 1 минуты',
        (60,299): '1-5 минут',
        (300,1199): '5-20 минут',
        (1200, session_durations.max()): 'более 20 минут'
        }

for key in duration_types.keys():
    durations_df.loc[len(durations_df.index)] = (
        [duration_types[key], duration_values[(duration_values['value'] >= key[0]) &
        (duration_values['value'] <= key[1])]['n_values'].sum()])
durations_df['type'] = durations_df.apply(
    lambda x: x['type'] + f' ({round(x.n_values / durations_df.n_values.sum() * 100, 2)}%)', axis=1)
    
durations_df
```

```python
plt.figure(figsize=(12, 6))
plt.grid()
plt.title('Продолжительность пользовательских сессий')
plt.ylabel('Количество сессий')
plt.bar(x=durations_df['type'], height=durations_df['n_values'])
plt.show()
```

Только **11%** сессий являются короткими. В остальных **87%** случаев пользователи проводят на сайте не меньше 1 минуты. А **13%** сессий длятся больше 20 минут. Это свидетельствует о высокой степени вовлеченности пользователей.

Для расчета средней продолжительности сессии подготовим гистограмму по всем значениям.

### Retention Rate по когортам

Проведем когортный анализ коэффициента удержания пользователей. Для начала оценим ежемесячное количество новых пользователей (размеры когорт).

Здесь мы можем наблюдать схожую с метриками DAU/WAU/MAU сезонность: в летние месяцы, а также во второй половине весны приток новых пользователей минимален. С октября по март - "высокий" сезон с пиком в октябре-декабре.

Выявим средний Retention Rate по всем когортам в разные периоды их жизни: со 2-го по конец 8-го месяца.

```python
visits_cohorts = visits.groupby(['cohort_month','cohort_lifetime']).agg({'uid':'nunique', 'cohort_size':'mean'}).reset_index()
visits_cohorts = visits_cohorts.rename(columns=({'uid':'visitors_per_month'}))
visits_cohorts['retention_rate'] = visits_cohorts['visitors_per_month'] / visits_cohorts['cohort_size']

cohorts_average = pd.DataFrame(columns=['cohort_lifetime', 'mean_retention'])

for i in range(1,8):
    mean_retention = visits_cohorts.query('cohort_lifetime == @i').groupby('cohort_month')['retention_rate'].mean().mean()
    cohorts_average.loc[len(cohorts_average.index)] = [i, round(mean_retention*100, 2)]

display(cohorts_average)
    
plt.figure(figsize=(12, 6))
plt.grid()
plt.title('Усредненный Retention Rate, все когорты')
plt.xlabel('Месяц жизни когорты')
plt.ylabel('Retention Rate')
plt.bar(x=cohorts_average['cohort_lifetime'], height=cohorts_average['mean_retention'])
plt.show()
```

Самой "выносливой" за весь рассматриваемый период оказалась самая старая когорта пользователей, которые начали пользоваться сайтом **в июне 2017 года** - даже на 12-й месяц существования когорты **4.5%** пользователей продолжают посещать сайт Яндекс.Афиши. Ни одна когорта (из просуществовавших от 2 и более месяцев) не смогла продемонстрировать подобный результат по состоянию на финальный месяц исследуемого периода.

Если говорить про общие тенденции, то пользователи когорт, сформировавшихся **с июня по сентябрь**, гораздо более склонны к повторному посещению ресурса, чем пользователи из других когорт.

### Время первой покупки

Изучим, сколько времени проходит с первого посещения сайта до совершения покупки.

```python
first_visit = visits.groupby(['uid'])['start_ts'].min()
first_visit.name = 'first_visit'
orders = orders.merge(first_visit, on='uid')
orders['time_to_decide'] = orders['buy_ts'] - orders['first_visit']
display(orders['time_to_decide'].describe())
```

Даже без графика видно, что в данных присутствует правосторонняя асимметрия - 4-й квартиль в диапазоне от 38 дней до почти года. В отличие от ситуации с пользовательскими сессиями, в которых максимальные значения были очевидными выбросами, в данном случае у нас нет оснований полагать, что высокие значения являются мусорными, поэтому мы оставим их в датафрейме, а в качестве меры средней тенденции возьмем медиану.

Медианное время на принятие решение о покупке составляет **3 часа 26 минут**.

Теперь разобьем покупателей на категории по времени принятия решения - в течение 1 часа / дня / недели / месяца / больше месяца - и оценим доли этих категорий.

```python
decision_speed_uid = (orders['time_to_decide'].value_counts().reset_index()
    .rename(columns = {'index':'value', 'session_duration_sec': 'n_values'}))

decision_speed = {
    (np.timedelta64(0, 's'), np.timedelta64(1, 'h')): 'В течение 1 часа',
    (np.timedelta64(1, 'h'), np.timedelta64(1, 'D')): 'От 1 до 24 часов',
    (np.timedelta64(1, 'D'), np.timedelta64(1, 'W')): 'От 1 до 7 дней',
    (np.timedelta64(1, 'W'), np.timedelta64(1, 'M')): 'От недели до месяца',
    (np.timedelta64(1, 'M'), orders['time_to_decide'].max()): 'Более 1 месяца'
}

decision_speed_df = pd.DataFrame(columns = ['type', 'n_values'])

for key in decision_speed.keys():
    decision_speed_df.loc[len(decision_speed_df.index)] = (
        [decision_speed[key], orders[(orders['time_to_decide'] >= key[0]) &
        (orders['time_to_decide'] <= key[1])]['uid'].nunique()])
    
decision_speed_df['type'] = decision_speed_df.apply(
    lambda x: x['type'] + f' ({round(x.n_values / decision_speed_df.n_values.sum() * 100, 2)}%)', axis=1)

display(decision_speed_df)

plt.figure(figsize=(12, 6))
plt.grid()
plt.title('Время на принятие решение о первой покупке')
plt.ylabel('Количество пользователей')
plt.bar(x=decision_speed_df['type'], height=decision_speed_df['n_values'])
plt.xticks(rotation=45)
plt.setp(plt.gca().xaxis.get_majorticklabels(), ha='right')
plt.show()
```

Больше половины (**54%**) покупателей совершает свою первую покупку в течение 1 часа, **83%** пользователей - в течение месяца. Однако доля пользователей, у которых уходит больше 1 месяца на принятие решения о покупке - также достаточно велика (**17%**). Поэтому весьма целесообразны мероприятия по повторному привлечению пользователей, еще не совершивших покупку.

### Пожизненная ценность клиента (LTV)

Оценим средний LTV в разрезе когорт.

```python
revenue_by_cohort = orders.groupby(['cohort_month','cohort_lifetime'])['revenue'].sum()
revenue_by_cohort.name = 'revenue_by_cohort'
orders = orders.merge(revenue_by_cohort, on=['cohort_month','cohort_lifetime'])

margin_rate = 1
''' несмотря на то что при маржинальности 100% прибыль де факто равна выручке,
для порядка все равно выведем показатель прибыли в отдельный столбец'''

orders['gp_by_cohort'] = orders['revenue_by_cohort'] * margin_rate
orders['gp_per_buyer'] = (orders['gp_by_cohort'] / orders['cohort_size']).round(2)
ltv_by_cohort = orders.pivot_table(index = 'cohort_month', columns = 'cohort_lifetime', values = 'gp_per_buyer', aggfunc = 'mean')
ltv_by_cohort = ltv_by_cohort.cumsum(axis=1)
ltv_by_cohort
```

```python
sns.set(style='white')
plt.figure(figsize=(12, 6))
plt.title('LTV по когортам')
sns.heatmap(ltv_by_cohort, annot=True, fmt='.2f', linewidths=1,
            linecolor='gray', yticklabels = [x.strftime('%Y-%m') for x in ltv_by_cohort.index])
plt.ylabel('Первый визит')
plt.xlabel('Возраст когорты, мес.')
plt.show()
```

Наилучший результат с точки зрения LTV демонстрируют когорты **с июня по сентябрь** включительно, причем июнь и сентябрь наиболее ценны. Начиная с октября динамика показателя существенно снижается. Единственным исключением является **декабрь** с его LTV на уровне июля-августа.

Феноменальные показатели июня и сентября, вероятно, могут быть связаны с формированием потребительской привычки в начале сезона: летнего (а также каникулярного) и осеннего (а также учебного). Так или иначе, многие рассмотренные метрики показывают нам, что июнь и сентябрь заслуживают особого внимания со стороны отдела маркетинга.

Поскольку CAC является все-таки вторичной метрикой, а первичен - именно финансовый результат, который мы получаем взамен на инвестиции, целесообразность применения того или иного источника в разные временные периоды мы будем оценивать с точки зрения ROMI.

### Коэффициент возврата маркетинговых инвестиций (ROMI)

Оценим общий ROMI по всем типам источников.

```python
revenue_per_source = orders_sources.groupby('source_id')['revenue'].sum().reset_index()
source_info = source_info.merge(revenue_per_source, on = 'source_id')
margin_rate = 1.0
source_info['gp'] = source_info['revenue'] * margin_rate
source_info['romi'] = round(source_info['gp'] / source_info['costs'], 2)
display(source_info[['source_id', 'costs', 'gp', 'romi']])

plt.figure(figsize=(12, 6))
plt.grid()
plt.title('Общий ROMI по типам источников')
plt.xlabel('Тип источника')
plt.bar(x = source_info['source_id'].astype('str'), height=source_info['romi'], align='center')
plt.show()
```

Наиболее эффективен источник **1** - за исследуемый период он полностью окупился и принес дополнительно **49%** прибыли. На втором месте источник **2** с ROMI равным **10%**.

Источники **5** и **9** принесли незначительную прибыль, источник **4** - 7% убытка.

Источники **3** и **10** приносят компании серьезные убытки - их ROMI **0.39** и **0.76** соответственно. 

Оценим средний ROMI по когортам в разные периоды их жизни в разрезе источников.

```python
orders_sources_cohorts = orders_sources.groupby(['source_id', 'cohort_lifetime'])['revenue'].sum().reset_index()
romi_by_lifetime = orders_sources_cohorts.merge(source_info[['source_id', 'costs']], on='source_id')
romi_by_lifetime['romi'] = romi_by_lifetime['revenue'] / romi_by_lifetime['costs']
romi_by_lifetime_pivot = romi_by_lifetime.pivot_table(index='cohort_lifetime',
                                                      columns='source_id', values='romi')
romi_by_lifetime_pivot.fillna(0, inplace=True)
romi_by_lifetime_pivot = romi_by_lifetime_pivot.cumsum(axis=0)

sns.set(style='white')
plt.figure(figsize=(12, 6))
plt.title('Средний накопительный ROMI по типам источников')
sns.heatmap(romi_by_lifetime_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.ylabel('Возраст когорты, мес.')
plt.xlabel('Тип источника')
plt.show()
```

Источник **1** в среднем окупается на 2-й месяц жизни когорт и демонстрирует хорошую динамику вплоть до финального исследуемого месяца. Источник **2** - начинает приносить прибыль на 6-й месяц. Источникам **9** и **5** потребовалось **7** и **8** месяцев соответственно.

Вложения в источнике **3** и **10**, судя по динамике (0.08 и 0.04 "дельты" за 12 месяцев), не окупятся никогда.

Теперь посчитаем общий ROMI по когортам за рассматриваемый период.

```python
orders_overall_pivot = orders_sources.pivot_table(index = 'cohort_month', columns = 'cohort_lifetime', values = 'revenue', aggfunc='sum')
orders_overall_pivot = orders_overall_pivot.cumsum(axis=1)
costs['month'] = costs['month'].astype('datetime64[M]')
costs_by_cohorts = costs
costs_by_cohorts = costs.groupby('month')['costs'].sum()

romi_cohorts = orders_overall_pivot

for column in romi_cohorts:
    romi_cohorts[column] = romi_cohorts[column] / costs_by_cohorts

sns.set(style='white')
plt.figure(figsize=(12, 6))
plt.title('Общий ROMI по когортам')
sns.heatmap(romi_cohorts, annot=True, fmt='.2f', linewidths=1, linecolor='gray',
            yticklabels = [x.strftime('%Y-%m') for x in romi_cohorts.index])
plt.ylabel('Первый визит')
plt.xlabel('Возраст когорты, мес.')
plt.show()
```

С точки зрения возврата маркетинговых инвестиций стоит обратить внимание на следующие когорты: **2017-06** (ROMI **1.33** по итогам 11 месяцев), **2017-07**, **2017-09** (самый высокий накопительный ROMI среди всех когорт, причем только на 8-й месяц) и **2017-12** (к сожалению, у нас нет данных о дальнейшей судьбы когорты, но по динамике можно ожидать от нее хороший финансовый результат).

К категории однозначно плохих когорт можно отнести **2018-01** и **2018-02**.

Для формирования полного представления об эффективности источников также целесообразно оценить эффективность каждого источника в разные временные периоды. Чтобы иметь выборку адекватного объема, мы возьмем для анализа когорты со сроком жизни 6 месяцев.

```python
source_cohort_6months = orders_sources[orders_sources['cohort_lifetime'] == 5]['cohort_month'].unique()
orders_6months = orders_sources[(orders_sources['cohort_month'].isin(source_cohort_6months)) & (orders_sources['cohort_lifetime'] <= 5)].pivot_table(index='cohort_month', columns='source_id', values='revenue', aggfunc='sum')
costs_6months = costs.query('month in @source_cohort_6months')
costs_6months = costs_6months.rename(columns=({'month':'cohort_month'}))
costs_6months = costs_6months.pivot_table(index='cohort_month', columns='source_id', values='costs', aggfunc='sum')
romi_cohorts_overall = orders_6months / costs_6months

romi_deviation = pd.DataFrame()

for column in romi_cohorts_overall:
    romi_deviation[column] = round((romi_cohorts_overall[column] - romi_cohorts_overall[column].mean()) / romi_cohorts_overall[column].mean() * 100, 2)

for column in romi_deviation.columns:
    romi_deviation[column] = romi_deviation[column].apply(lambda x: f'(+{str(x)}%)' if x>0 else f'({str(x)}%)')
    
sns.set(style='white')
plt.figure(figsize=(12, 6))
plt.title('ROMI для 6-месячных когорт и отклонение от среднего ROMI для типа источника')
sns.heatmap(romi_cohorts_overall, annot=True, annot_kws={'va':'bottom', 'size':'14'},
            fmt='.2f', linewidths=1, linecolor='gray',)
sns.heatmap(romi_cohorts_overall, annot=romi_deviation, fmt='1',
            annot_kws={'va':'top', 'size':'10'}, linewidths=1,
            linecolor='gray', cbar=False,
            yticklabels = [x.strftime('%Y-%m') for x in romi_cohorts_overall.index])
plt.ylabel('Первый визит')
plt.xlabel('Тип источника')
plt.yticks(rotation=0) 
plt.show()
```

Как мы видим, некоторые источники демонстрируют экстраординарные результаты в конкретных когортах - вероятно, они лучше всего попадают в целевую аудиторию. В частности:

 - Источник **1** наиболее эффективен в когортах **2017-06** (ROMI за полгода на **107%** выше, чем в среднем по остальным рассмотреным когортам) и **2017-07**(прирост **44%** от среднего результата)
 - Источник **2** эффективен в когорте **2017-12** (прирост **134%**)
 - Источник **5** эффективен в когорте **2017-09** (прирост **183%** и самый высокий ROMI среди рассмотренных комбинаций источник-когорта)
 
Прежде чем выдавать рекомендацию о точечном применении источников, мы должны убедиться в достаточном объеме выборки.

```python
efficient_sources = {
    1: ['2017-06', '2017-07'],
    2: ['2017-12'],
    5: ['2017-09']
}

for source in efficient_sources.keys():
    efficient_cohorts = efficient_sources[source]
    for cohort in efficient_cohorts:
        result = orders_sources.query('cohort_month == @cohort & source_id == @source')['uid'].nunique()
        cohort_size = orders_sources.query('cohort_month == @cohort')['uid'].nunique()
        ratio = round(result / cohort_size * 100, 2)
        print(f'Тип источника {source}, когорта {cohort}, количество пользователей {result} ({ratio}% от всей когорты)')
```

Как видим, топовые источники дают когортам от **8%** до **22%** аудитории. Можно смело вносить источники в рекомендации.

Типы используемых источников:

 - Рекомендуется отказаться от применения источников **3** и **10**, либо кардинально изменить подход к их применению

Распределение маркетингового бюджета:
 - Рекомендуется значительно увеличить расходы на привлечение пользователей в период **с июня по сентябрь**. Сделать это можно за счет снижения инвестиций в маркетинг в период **с января по апрель** - аудитория, которая приходит в эти месяцы, гораздо менее активна и приносит меньше прибыли, чем "летне-осенние" пользователи.

Ввиду очень высоких показателей Retention, количества заказов на 1 покупателя, LTV и ROMI рекомендуется сделать точечный акцент на привлечении пользователей в следующих когортах:

 - **Сентябрь**. Особенно эффективен источник **№5** (прирост **183%**)
 - **Декабрь**. Особенно эффективен источник **№2** (прирост **134%**)
